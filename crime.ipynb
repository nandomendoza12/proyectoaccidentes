```python
#Convertir Date Rptd a formato datetime
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%Y-%m-%d', errors='coerce')
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%Y-%m-%d', errors='coerce')
```

```python
# Paso 1: Extraer los primeros 3 dígitos de TIME OCC
col = 'TIME OCC'

# Convertir a string
s = df[col].astype(str).str.strip()

# Extraer primeros 3 dígitos y crear columna
df['TIME_OCC_3DIGITS'] = s.str[:3]

print('Paso 1: Primeros 3 dígitos de TIME OCC')
print(df[['TIME OCC', 'TIME_OCC_3DIGITS']].head(20))
print(f"\nValores únicos en TIME_OCC_3DIGITS: {df['TIME_OCC_3DIGITS'].nunique()}")

# Paso 2: Convertir a decimal e interpretar como horas:minutos
# Insertar coma después del primer dígito: 315 -> 3,15
df['TIME_DECIMAL'] = (df['TIME_OCC_3DIGITS'].str[:1].astype(float) + 
                      df['TIME_OCC_3DIGITS'].str[1:].astype(float) / 100)

# Paso 3: Convertir decimal a HH:MM
# Parte entera = horas
df['TIME_HOUR'] = df['TIME_DECIMAL'].astype(int)

# Parte decimal × 60 = minutos (0.15 × 60 = 9, 0.75 × 60 = 45)
df['TIME_MINUTE'] = ((df['TIME_DECIMAL'] - df['TIME_HOUR']) * 60).round(0).astype(int)

# Paso 4: Crear columna formateada HH:MM
df['TIME_OCC_FORMATTED'] = df['TIME_HOUR'].astype(str).str.zfill(2) + ':' + df['TIME_MINUTE'].astype(str).str.zfill(2)

# Resumen
print('\nPaso 2-4: Conversión decimal a HH:MM')
print(f"Total de filas: {len(df)}")
print(f"Conversiones exitosas: {df['TIME_OCC_FORMATTED'].notna().sum()}")

print('\nEjemplos de conversiones:')
print(df[['TIME OCC', 'TIME_OCC_3DIGITS', 'TIME_DECIMAL', 'TIME_HOUR', 'TIME_MINUTE', 'TIME_OCC_FORMATTED']].head(15))

print('\nDistribución de horas:')
print(df['TIME_HOUR'].value_counts().sort_index())
```

```python
#Vamos verifcar si tiene nulos todas las columnas vamos usar una funcion para que sea mas rapido
def verificacion_nulos(df2):
    """
    Función para verificar valores nulos en un DataFrame
    Retorna un resumen detallado de nulos por columna
    """
    print('='*70)
    print('VERIFICACIÓN DE VALORES NULOS')
    print('='*70)
    
    # Contar nulos por columna
    nulos = df.isnull().sum()
    total_filas = len(df)
    porcentaje = (nulos / total_filas * 100).round(2)
    
    # Crear DataFrame de resumen
    resumen_nulos = pd.DataFrame({
        'Columna': nulos.index,
        'Nulos': nulos.values,
        'Total': total_filas,
        'Porcentaje (%)': porcentaje.values
    })
    
    # Filtrar solo columnas con nulos
    resumen_nulos_con_nulos = resumen_nulos[resumen_nulos['Nulos'] > 0].sort_values('Porcentaje (%)', ascending=False)
    
    print(f"\nTotal de filas: {total_filas}")
    print(f"Total de columnas: {len(df.columns)}\n")
    
    if len(resumen_nulos_con_nulos) == 0:
        print("✓ ¡EXCELENTE! NO HAY VALORES NULOS EN NINGUNA COLUMNA")
    else:
        print("⚠ COLUMNAS CON VALORES NULOS:")
        print(resumen_nulos_con_nulos.to_string(index=False))
    
    print('\n' + '='*70)
    return resumen_nulos

# Cargar el dataset limpio
df2 = pd.read_csv('Crimenes2023-2025_limpio.csv')

# Ejecutar verificación
resumen = verificacion_nulos(df2)
```

```python
# CATEGORIZAR PERÍODOS DEL DÍA - BASADO EN APROXIMACIÓN DE HORA
# ⚠ NOTA IMPORTANTE: Como TIME OCC no tiene hora exacta (solo primeros 3 dígitos),
# esta categorización es APROXIMADA y solo para fines de análisis en Power BI

# Paso 1: Extraer la hora aproximada en una columna nueva
df2['Hour_OCC'] = df2['TIME_OCC_FORMATTED'].str[:2].astype(int)

# Paso 2: Crear función para categorizar períodos
def categorizar_periodo(hour):
    """
    Categoriza la hora APROXIMADA en períodos del día:
    - Madrugada: 00:00 - 05:59
    - Mañana: 06:00 - 11:59
    - Tarde: 12:00 - 17:59
    - Noche: 18:00 - 23:59
    
    ⚠ NOTA: Esta es una categorización APROXIMADA basada en los 
    primeros 3 dígitos de TIME OCC, NO es una hora exacta.
    """
    if pd.isna(hour):
        return 'Unknown'
    elif 0 <= hour < 6:
        return 'Madrugada'
    elif 6 <= hour < 12:
        return 'Mañana'
    elif 12 <= hour < 18:
        return 'Tarde'
    else:  # 18 <= hour < 24
        return 'Noche'

# Paso 3: Aplicar la función
df2['Period_Day'] = df2['Hour_OCC'].apply(categorizar_periodo)

# Paso 4: Mostrar resultados
print('='*70)
print('CLASIFICACIÓN APROXIMADA DE INCIDENTES POR PERÍODO DEL DÍA')
print('='*70)
print('\n⚠ IMPORTANTE:')
print('   Como TIME OCC no proporciona hora exacta, esta categorización')
print('   es APROXIMADA basada en los primeros 3 dígitos del número original.')
print('   Usar solo para análisis general en Power BI.')

print('\nEjemplos de clasificación:')
print(df2[['TIME_OCC_FORMATTED', 'Hour_OCC', 'Period_Day']].head(15))

print('\n\nDistribución de incidentes por período (APROXIMADA):')
period_counts = df2['Period_Day'].value_counts()
print(period_counts)

print('\n\nPorcentaje por período:')
period_pct = (df2['Period_Day'].value_counts() / len(df2) * 100).round(2)
for period, pct in period_pct.items():
    print(f"  {period}: {pct}%")

print('\n' + '='*70)
print('✓ Columnas creadas:')
print('  - Hour_OCC: Hora aproximada (0-23)')
print('  - Period_Day: Categoría APROXIMADA del período')
print('  - TIME_OCC_FORMATTED: Valor original intacto')
print('\n⚠ Estas columnas son APROXIMADAS. No usar para análisis de precisión.')
print('='*70)

# Guardar dataset actualizado
df2.to_csv('Crimenes2023-2025_limpio.csv', index=False)
print('\n✓ Dataset guardado')
```
